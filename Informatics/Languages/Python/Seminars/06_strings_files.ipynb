f-—Å—Ç—Ä–æ–∫–∏ –ø–æ–¥–¥–µ—Ä–∂–∏–≤–∞—é—Ç —Ñ–æ—Ä–º–∞—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ —á–∏—Å–µ–ª:

–¢–∞–∫–∂–µ –º–æ–¥—É–ª—å –ø–æ–∑–≤–æ–ª—è–µ—Ç —Å–æ–∑–¥–∞–≤–∞—Ç—å —à–∞–±–ª–æ–Ω—ã c –ø–æ–º–æ—â—å—é `Template` –∏ –º–µ—Ç–æ–¥–∞ `substitute`:

```python
from string import Template

s = Template('$who likes $what')
s.substitute(who='tim', what='kung pao')
```

–ï—Å–ª–∏ –≤ —Å–ª–æ–≤–∞—Ä–µ –æ—Ç—Å—É—Ç—Å—Ç–≤—É–µ—Ç –æ–¥–∏–Ω –∏–∑ –∫–ª—é—á–µ–π, —Ç–æ –≤—ã–ª–µ–∑–µ—Ç –æ—à–∏–±–∫–∞.

```python
d = dict(who='tim')
Template('$who likes $what').substitute(d)
```

–ß—Ç–æ–±—ã —ç—Ç–æ–≥–æ –Ω–µ –±—ã–ª–æ, –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –º–µ—Ç–æ–¥ `safe_substitute`:

```python
Template('$who likes $what').safe_substitute(d)
```

–ü–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ —Ç–æ, –∫–∞–∫ –≤—ã–≥–ª—è–¥–∏—Ç –∫–∞–∂–¥—ã–π –±–∞–π—Ç –≤ –æ—Ç–¥–µ–ª—å–Ω–æ—Å—Ç–∏:

–ö–∞–∫–∏–µ –∫–æ–¥–∏—Ä–æ–≤–∫–∏ –º–æ–∂–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å? –†–∞–∑–Ω—ã–µ:

```python
string = 'El Ni√±o'

for codec in ['latin_1', 'utf_8', 'utf_16', 'cp437']:
    encoded = string.encode(codec)
    print(codec, encoded.decode(codec), encoded, sep='\t\t')
```

–ú–µ—Ç–æ–¥ `f.read(n)` —á–∏—Ç–∞–µ—Ç `n` —Å–∏–º–≤–æ–ª–æ–≤ (–∫–æ–≥–¥–∞ —Ñ–∞–π–ª –±–ª–∏–∑–∏—Ç—Å—è –∫ –∫–æ–Ω—Ü—É –∏ –ø—Ä–æ—á–∏—Ç–∞—Ç—å –∏–º–µ–Ω–Ω–æ `n` —Å–∏–º–≤–æ–ª–æ–≤ —É–∂–µ –Ω–µ–≤–æ–∑–º–æ–∂–Ω–æ, —á–∏—Ç–∞–µ—Ç –º–µ–Ω—å—à–µ; –≤ —Å–∞–º—ã–π –ø–æ—Å–ª–µ–¥–Ω–∏–π —Ä–∞–∑ –æ–Ω —á–∏—Ç–∞–µ—Ç 0 —Å–∏–º–≤–æ–ª–æ–≤ –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç `''`). –ü—Ä–æ—á–∏—Ç–∞–µ–º —Ñ–∞–π–ª –ø–æ 2 —Å–∏–º–≤–æ–ª–∞.

```python
with open('text.txt') as f:
    while True:
        c = f.read(2)
        if c == '':
            break
        else:
            print(c.__repr__())
```

–ï—Å–ª–∏ —Ö–æ—Ç–∏–º –ø—Ä–æ—á–∏—Ç–∞—Ç—å —Ñ–∞–π–ª –ø–æ—Å—Ç—Ä–æ—á–Ω–æ, —Ç–æ –ø–æ–ª—å–∑—É–µ–º—Å—è –º–µ—Ç–æ–¥–æ–º `readline`:

```python
with open('text.txt') as f:
    count = 0
    while True:
        s = f.readline()
        if s == '':
            break
        else:
            print(f'count: {count}, {s.__repr__()}')
            count += 1
```

```python
with open('text.txt') as f:
    s1 = f.readline()[:-1]
    s2 = f.readline()[:-1]
    s3 = f.readline()[:-1]

print(s1, s2, s3, sep='\n')
```

–¢–µ–ø–µ—Ä—å –ø–æ—Å–º–æ—Ç—Ä–∏–º, —á–µ–º –∂–µ –æ–ø–µ—Ä–∞—Ç–æ—Ä `with` –ª—É—á—à–µ, —á–µ–º –ø–∞—Ä–∞ `open` - `close`.

```python
def a(name):
    global f
    f = open(name)
    s = f.readline()
    n = 1 / 0
    f.close()
    return s
```

–ü—Ä–æ–∏–∑–æ—à–ª–æ –∏—Å–∫–ª—é—á–µ–Ω–∏–µ, –º—ã –ø–æ–∫–∏–Ω—É–ª–∏ —Ñ—É–Ω–∫—Ü–∏—é –¥–æ —Å—Ç—Ä–æ—á–∫–∏ `close`, –∏ —Ñ–∞–π–ª –Ω–µ –∑–∞–∫—Ä—ã–ª—Å—è.

```python
def a(name):
    global f
    with open(name) as f:
        print(f'File is opened: {not f.closed}')
        s = f.readline()
        n = 1 / 0
    return s
```

–°–æ–∑–¥–∞–¥–∏–º —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –∫–æ–ø–∏—Ä—É–µ—Ç —Å—Ç–∞—Ä—ã–π —Ç–µ–∫—Å—Ç–æ–≤—ã–π —Ñ–∞–π–ª –≤ –Ω–æ–≤—ã–π. –ï—Å–ª–∏ —Å—Ç—Ä–æ–∫–∏ –Ω—É–∂–Ω–æ –∫–∞–∫-–Ω–∏–±—É–¥—å –æ–±—Ä–∞–±–æ—Ç–∞—Ç—å, –≤ –ø–æ—Å–ª–µ–¥–Ω–µ–π —Å—Ç—Ä–æ—á–∫–µ –≤–º–µ—Å—Ç–æ `line` –±—É–¥–µ—Ç —Å—Ç–æ—è—Ç—å —á—Ç–æ-–Ω–∏–±—É–¥—å –≤—Ä–æ–¥–µ `f(line)`:

```python
def copy(old_name, new_name):
    with open(old_name) as old, open(new_name, 'w') as new:
        for line in old:
            new.write(line)
```

```python
def copy(old_name, new_name):
    with open(old_name) as old:
        with open(new_name, 'w') as new:
            for line in old:
                new.write(line)
```

```python
copy('text.txt', 'newtext.txt')
```

**–†–µ–≥—É–ª—è—Ä–Ω—ã–µ –≤—ã—Ä–∞–∂–µ–Ω–∏—è** - —Ñ–æ—Ä–º–∞–ª—å–Ω—ã–π —è–∑—ã–∫ –ø–æ–∏—Å–∫–∞ –∏ –æ—Å—É—â–µ—Å—Ç–≤–ª–µ–Ω–∏—è –º–∞–Ω–∏–ø—É–ª—è—Ü–∏–π —Å –ø–æ–¥—Å—Ç—Ä–æ–∫–∞–º–∏ –≤ —Ç–µ–∫—Å—Ç–µ, –æ—Å–Ω–æ–≤–∞–Ω–Ω—ã–π –Ω–∞ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–∏ –º–µ—Ç–∞—Å–∏–º–≤–æ–ª–æ–≤ (`. ÀÜ $ * + ? { } [ ] | ( )`)

–í Python –¥–ª—è —Ä–∞–±–æ—Ç—ã —Å —Ä–µ–≥—É–ª—è—Ä–Ω—ã–º–∏ –≤—ã—Ä–∞–∂–µ–Ω–∏—è–º–∏ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç –º–æ–¥—É–ª—å `re`.

–î–ª—è –ø–æ–∏—Å–∫–∞ –≤—Å–µ—Ö –Ω–µ–ø–µ—Ä–µ—Å–µ–∫–∞—é—â–∏—Ö—Å—è –≤—Ö–æ–∂–¥–µ–Ω–∏–π —Ä–µ–≥—É–ª—è—Ä–Ω–æ–≥–æ –≤—ã—Ä–∞–∂–µ–Ω–∏—è –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è `re.findall`:

```python
import re

string = '__abc__acc__abcd__a6c__a66c'
```

```python
re.findall(r'abc', string)
```

```python
re.findall(r'a\dc', string)
```

```python
re.findall(r'a\w+c', string)
```

–ú—ã —Ç–∞–∫–∂–µ –º–æ–∂–µ–º —Å–æ—Ö—Ä–∞–Ω–∏—Ç—å –ø–∞—Ç—Ç–µ—Ä–Ω –∏ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –µ–≥–æ –Ω–µ—Å–∫–æ–ª—å–∫–æ —Ä–∞–∑:

```python
text = u'Fran√ßais z≈Çoty √ñsterreich'
pattern = r'\w+'
ascii_pattern = re.compile(pattern, re.ASCII)
unicode_pattern = re.compile(pattern)
```

```python
text
```

```python
pattern
```

```python
list(ascii_pattern.findall(text))
```

```python
list(unicode_pattern.findall(text))
```
